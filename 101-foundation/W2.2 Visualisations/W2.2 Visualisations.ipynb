# Visualisations with Matplotlib & Seaborn

Data Visualisation is a massive field these days and it's a fundamental part of the Data Science and Machine Learning process. 
<br> <br>
But they are not new in any way shape of form. The idea goes back centuries and millenia, because in essence, they are attempts to make complicated information into digestible and easily readable graphics. Maps are a classic example: they take a lot of the information about the landscape around us and simplify it to the extent that it can be read on 2D paper. 
<br><br>
All visualisations follow that principle to some extent - they take complicated and messy data and make it easier for humans to access or understand it. 


## Rules of thumb when Visualising

1) **Value & Purpose:** They must fulfill a purpose and add value by making data and the patterns within it easier to read
<br> <br>
2) **Simplicity:** when choosing between visualisations representing the same purpose, the one that is easier to read is always better
<br> <br>
3) **Continuation:** You can slice and dice the data in many different ways. Normally, you wil use visualisations when trying to explain the dataset or present your analysis to someone, perhaps your boss. Out of the infinite types of visualisations that you can present, choose carefully the one that you think fits your narrative best, but is also representative of the truth. 
<br> <br>
4) **Audience:** Visualisations can get complicated, and you must keep your audience in mind while presenting. e.g. if you are presenting to a group of non-technical people, it would be best to keep it simple.

## Histograms

Histograms are a great way to see how our columns are distributed and shaped. Let's see what we can learn about our dataset this way.<br>
### Uses
1) Data Exploration
<br>
2) Seeing the spread of your columns so you can check where to start looking etc

```python
# First we will import matplotlib

import matplotlib.pyplot as plt
```

```python
# Let's first learn how to use a histogram on just one variable: satisfaction_level

plt.figure(figsize=(10,7)) # Define how big you want the graph to be (in inches)
plt.hist(HR['satisfaction_level']) # Pass the column that you want to run the histogram on
plt.title('Satisfaction Level Spread') 
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')
plt.show()
```

```python
# What if you wanted to compare two variables on one plot?
plt.figure(figsize=(20,10))
plt.hist(HR['satisfaction_level'], label = 'Satisfaction') # Pass the column that you want to run the histogram on
plt.hist(HR['last_evaluation'],color='r', label = 'Last Evaluation')
plt.title('Satisfaction Level Spread') 
plt.legend()
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')
plt.show()
```

Wow, that's really ugly. 

Although useful for a glance, that's not exactly presentable. We'll see soon how to use seaborn to make that prettier

```python
def histogram(col):
    plt.figure(figsize=(10,7))
    plt.hist(HR[col])
    plt.title('Histogram for ' + col)
    plt.xlabel('Column Value')
    plt.ylabel('Frequency')
    plt.show()    
```

```python
for col in range(len(list(HR))-2):
    histogram(list(HR)[col])
```

Now we have a better idea of how our columns our spaced. Let's see what we notice from this dataset. 

## Scatter Plots

Scater plots are also something you end up using a lot. They are mostly used to see how two variables are interconnected or spaced out in relation to each other.

Let's say you want to see how the relationship between satisfaction level and numbers of hours worked looks like. This is how you would scatter them:

```python

plt.figure(figsize=(20,15))
plt.scatter(HR['satisfaction_level'],HR['average_montly_hours'])
plt.xlabel('Satisfaction Level', fontsize=20)
plt.ylabel('Average Monthly Hours Worked', fontsize=20)
plt.title('Satisfaction Level vs Average Monthly Hours Worked')
plt.show()
```

## Bar Charts
 I'm sure you all know what bar graphs are, but to reiterate, they're graphs used to graph categorical or discrete data. 
 
 Let's find out the department distributions of our employees. 

```python
# We're going to use this Counter function that comes in very handy when finding out how many times something is repeated

from collections import Counter

depts = Counter(HR['sales'])
print(depts)
```

```python
plt.figure(figsize=(15,10))
plt.bar(range(len(depts)), depts.values(),align='center')
plt.xticks(range(len(depts)), depts.keys(),rotation='vertical')
plt.title('Headcount by Departments')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
# Let's try make the distribution plots (Histograms) again but take it a step further
#First, we import seaborn

import seaborn as sns

# Set up the matplotlib figure
f, axes = plt.subplots(ncols=3, figsize=(15, 6))

# Graph Employee Satisfaction
sns.distplot(HR.satisfaction_level, kde=False, color="#800000", ax=axes[0]).set_title('Employee Satisfaction Level')

# Graph Employee Evaluation
sns.distplot(HR.last_evaluation, kde=False, color="b", ax=axes[1]).set_title('Employee Last Evaluation')

# Graph Employee Average Monthly Hours
sns.distplot(HR.average_montly_hours, kde=False, color="green", ax=axes[2]).set_title('Employee Average Monthly Hours')
plt.show()
```

```python
# Pretty simle and much more powerful right? Let's try something more
# What if I want to check the number of people leaving depending on their salary band?
# We can use seaborn pretty quickly to get this done by using this function called countplot

f, ax = plt.subplots(figsize=(15, 4))
sns.countplot(y="salary", hue='left', data=HR).set_title('Employee Turnover by Salary')
plt.show()
```

As you can see, we can do a lot more with less code! Let's move on to more advanced visualisations

```python
# What if we want to see the evaluations of employees who left, and who didn't leave?
# We can use this thing called a Kernel Density Plot
# In matplotlib, we would have to do a lot of wrangling with Pandas beforehand, but with Seaborn, it's pretty simple

fig = plt.figure(figsize=(20,8),)
ax=sns.kdeplot(HR.loc[(HR['left'] == 0),'last_evaluation'] , color='g',shade=True,label='Didn\'t Leave')
ax=sns.kdeplot(HR.loc[(HR['left'] == 1),'last_evaluation'] , color='b',shade=True, label='Left')
plt.xlabel('Evaluations')
plt.title('Employee\'s Evluations - Those who left vs didn\'t leave', fontsize=32)
plt.show()
```

```python
# A little more advanced, what if we wanted to see how the variables correlate to each other? 
# For those who don't know, correlation between two variables is how much the change in one explains the distribution
# or variance of the other. 
# Correlation is between the scale of -1 to 1: -1 being perfect negative correlation and 1 being perfect positive correlation

correlations_map= HR.corr()
plt.figure(figsize=(15,12))
sns.heatmap(correlations_map,annot=True,cbar=True,cmap="coolwarm_r")
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
# What if I want to see all combinations of variables scattered in a matrix?
# Essentially compare everything to each other to see what the relationships are like, depending on a hue of course

sns.set(style="darkgrid")
sns.pairplot(HR, hue="left")
plt.savefig('pairplot.png')
plt.show()
```

## Visualisation Tips for Presentations
Before we move on to you guys trying this on your own, here are some tips for using visualisations in presentations:

1) Relevant graphs only (depending on the presentation)
<br>
2) Always label the graphs with titles, axis names and legends when appropriate
<br>
3) Don't overcomplicate it. No one cares if you can make a 3-D Hexbin plot if no one can understand it
<br>
4) Don't ignore your graphs. If you have them in your presentations, talk about them and use them. Otherwise, don't include them.
<br>
5) They are there to help you explain your point. Make sure they add some value, or worse, aren't distracting
<br>
6) Give yourself time to make them: it's not as easy as it looks, especially with unclean data
